```python
import tikzplotlib
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv(\
                   '01-mean-flow/mean-flow129.dat', \
                   delim_whitespace= True)
```

```python
rad = data.radius
Mx  = data.M_x
Mt  = data.M_theta
```

```python
plt.plot(rad,Mx, label = 'M_x')
plt.plot(rad,Mt, label = 'M_t')
plt.plot(rad,(Mt**2+Mx**2)**0.5, label = 'M_total')
plt.legend()

plt.title('Mean Flow ')
plt.xlabel('Mach Number')
plt.ylabel('Radius')
tikzplotlib.save("MachDistPy.tex")
```
